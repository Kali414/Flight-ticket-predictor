```python
#Data preprocessing
dt_col=["date_of_journey","dep_time","arrival_time"]
num_col=["duration","total_stops"]
cat_col=["airline","source","destination","additional_info"]
##category=[col for col in X_train.columns if(col not in dt_cols ) and(col not in num_col)]
```

```python
#importing all the libraries
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

#column transformet is used to perform the set of function on each col
from sklearn.compose import ColumnTransformer
#for datetime things 
from feature_engine.datetime import DatetimeFeatures
```

```python
#creating of pipeline
num_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="mean")),
    ("scaler",StandardScaler())
])

cat_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("encoder",OneHotEncoder(sparse_output=False, handle_unknown="ignore"))
])

#feature extraction for time
time_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("extractor", DatetimeFeatures(features_to_extract=["hour","minute"], format="mixed")),
    ("scaler",StandardScaler())
])

#feature extraction for date
doj_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("extractor",DatetimeFeatures(features_to_extract=["month","week","day_of_week","day_of_month"],format="mixed")),
    ("scaler",StandardScaler())
])
```

```python
processor=ColumnTransformer(transformers=[
    ("num",num_pipe,num_col),
    ("cat",cat_pipe,cat_col),
    ("doj",doj_pipe,["date_of_journey"]),
    ("time",time_pipe,["dep_time","arrival_time"])
])
```

```python
processor.fit_transform(X_train)
```
